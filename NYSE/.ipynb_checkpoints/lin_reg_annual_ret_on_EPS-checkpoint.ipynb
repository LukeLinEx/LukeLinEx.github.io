```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
adjusted = pd.read_csv('./input/prices-split-adjusted.csv')
fundamentals = pd.read_csv('./input/fundamentals.csv', index_col=0)
securities = pd.read_csv('./input/securities.csv')
```

```python
def get_date(d):
    return dt.datetime.date(dt.datetime.strptime(d, '%Y-%m-%d'))

fundamentals['Period Ending'] = map(get_date, fundamentals['Period Ending'])
adjusted['date'] = map(get_date, adjusted['date'])
```

### To find symbols whose period are not close to a year

```python
def get_difference(seq_date):
    seq_date = sorted(map(lambda x: x.year*12+x.month, seq_date))
    start_= seq_date[:-1]
    end_  = seq_date[1:]
    return np.mean(map(lambda s, e: e - s, start_, end_))

avg_period_len = fundamentals.groupby('Ticker Symbol').agg({'Period Ending':get_difference})
avg_period_len.loc[(np.abs(avg_period_len -12)>1)['Period Ending'],:]
```

## Clean fundamentals

- period start is the last period ending
- use the earning per share attaching to the period ending (not period start) as the eps of the particular period
- remove the period missing eps

```python
tmp_s = fundamentals[['Ticker Symbol', 'Period Ending']].iloc[:-1].reset_index(drop=True)
tmp_e = fundamentals[['Ticker Symbol','Period Ending', 'Earnings Per Share']].iloc[1:].reset_index(drop=True)
tmp_s.columns.values[1] = 'Period Start'
tmp_e.columns.values[0] = 'symbol'
tmp = pd.concat([tmp_s, tmp_e], axis =1)
tmp = tmp.loc[tmp['Ticker Symbol'] == tmp['symbol'], ['symbol', 'Period Start', 'Period Ending',
       'Earnings Per Share']]
tmp = tmp.loc[map(lambda x: not x, tmp['Earnings Per Share'].isnull())]
tmp.head(100)
```

---

## Prepare data for linear regression

- combine adjusted and tmp (cleaned fundamentals)
- make sure the date (from adjusted) is in the within the range of period (excluding period start but including period ending)

```python
combine = pd.merge(adjusted[['symbol','date', 'close']],
                   tmp, on = 'symbol')

def check_date_in_rng(date_, period_s, period_e):
    return date_ > period_s and (period_e - period_s) >= (date_ - period_s)

combine = combine.loc[map(check_date_in_rng, 
                          combine['date'], 
                          combine['Period Start'], 
                          combine['Period Ending'])]

### Make sure that observations for each symbol is ordered by date
combines = combine.sort_values(['symbol', 'date'])

print combine.head()
print combine.tail()
```

```python
def get_return_in_period(x):
    return (x.iloc[-1]-x.iloc[0])/x.iloc[0]/len(x)

data = combine.groupby(['symbol', 'Period Start', 'Period Ending']).agg({'close':get_return_in_period,
                                                  'Earnings Per Share': 'first'})

data.head(15)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(data['Earnings Per Share'], data['close'], alpha=0.2)
```

```python
import statsmodels.api as sm

x = data[['Earnings Per Share']]
x = sm.add_constant(x)
model = sm.OLS(data['close'], x)
results = model.fit()
print results.summary()
```
