Let
- $X$ be data
- $L$ be loading vectors of PCA
- $Pc$ be the principal components
- $n$ be the number of rows
- $D$ be the covariance matrix $Pc$. So $D = Pc \times Pc^{T} / n $

$$
X L = Pc \ \rightarrow \ L^{T}X^{T}XL = Pc^{T} Pc = D \times n
$$

Therefore

$$
X \times D \times n = X \times Pc^{T} Pc
$$

So 

$$
 X = \frac{X Pc^{T}}{n} \times (Pc D^{-1}) = CoV(X, Pc) \times \text{scores}
$$

Below confirms:

```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
M = np.mean(X, axis=0)
X = X - M
X = np.matrix(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
S = np.matrix(pca.fit_transform(X))
```

```python
D = np.cov(S.T)
D[np.abs(D) <= 0.00001] = 0
D = np.matrix(D)
D
```

```python
np.round(
    ((X)*S.T)*S*D.I/150 + np.matrix(M),
    1
)
```
